# Avaliação

```python
import sqlite3
```

```python
name = 'escola.db'

script_alunos = """CREATE TABLE IF NOT EXISTS escola (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                nome TEXT NOT NULL,
                idade INTEGER NOT NULL,
                email TEXT NOT NULL
            );"""

try:
    with sqlite3.connect(name) as conn:

        cur = conn.cursor()
        cur.execute(script_alunos)
        conn.commit
        
        print('Criação da tabela bem sucedida')

except sqlite3.OperationalError as e:
    print('Criação da Tabela mal sucedida: ', e)


```

```python
sql = "INSERT INTO escola (nome, idade, email) VALUES (?,?,?)"

try:
    with sqlite3.connect(name) as conn:

        cur = conn.cursor() 
        cur.execute(sql, ('Joãozinho',78, 'Joãozinho@gmail' ))
        cur.execute(sql, ('Joãozão',2, 'Joãozão@gmail' ))
        cur.execute(sql, ('Jonas',34, 'Jonaz@gmail' ))
        cur.execute(sql, ('Jotaro',41, 'Jotaro@gmail' ))
        cur.execute(sql, ('Joabson',9, 'Joabson@gmail' ))

        conn.commit()

except sqlite3.OperationalError as e:
    print('Erro: ', e)
```

```python
sql = "SELECT * FROM escola"


try:
    with sqlite3.connect(name) as conn:
        cur = conn.cursor()
        cur.execute(sql)
        res = cur.fetchall()
        print(res)
except sqlite3.OperationalError as e:
    print("Erro: ", e)
```

```python
sql = "UPDATE escola SET nome = ? SET email = ? WHERE id = ?"

try:
    with sqlite3.connect(name) as conn:
             
        cur = conn.cursor()
        cur.execute(sql, ('Joviel','Joviel@gmail', 1))
        conn.commit()


except sqlite3.OperationalError as e:
    print("ERRO: ", e)
```

```python
sql = "DELETE FROM escola WHERE id = ?"

try:
    with sqlite3.connect(name) as conn:
            
        cur = conn.cursor()
        cur.execute(sql, (2,))
        conn.commit()
        


except sqlite3.OperationalError as e:
    print("ERRO: ", e)
```
